# 3. Calificación promedio de la tienda


```python
import pandas as pd

# Asignar nombres a las tiendas (si aún no lo has hecho)
tienda1["Tienda"] = "Tienda 1"
tienda2["Tienda"] = "Tienda 2"
tienda3["Tienda"] = "Tienda 3"
tienda4["Tienda"] = "Tienda 4"

# Unir todas las tiendas en un único DataFrame
todas_tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4])

# Calcular el promedio de calificaciones por tienda
calificacion_promedio = todas_tiendas.groupby("Tienda")["Calificación"].mean().round(2).reset_index()
calificacion_promedio.columns = ["Tienda", "Calificación Promedio"]  # Renombrar columnas

# Ordenar de mayor a menor satisfacción
calificacion_promedio = calificacion_promedio.sort_values("Calificación Promedio", ascending=False)

print("Calificación promedio por tienda:")
print(calificacion_promedio)
```

```python
import matplotlib.pyplot as plt

# Gráfico de barras
calificacion_promedio.plot(kind="bar", x="Tienda", y="Calificación Promedio",
                          title="Calificación Promedio por Tienda", legend=False)
plt.ylim(0, 5)  # Rango de calificaciones (1-5)
plt.ylabel("Calificación Promedio")
plt.xticks(rotation=0)
plt.show()
```

# 4. Productos más y menos vendidos

```python
import pandas as pd

# Asignar nombres de tienda si no están definidos
tienda1["Tienda"] = "Tienda 1"
tienda2["Tienda"] = "Tienda 2"
tienda3["Tienda"] = "Tienda 3"
tienda4["Tienda"] = "Tienda 4"

# Unir todas las tiendas
todas_tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4])

# Contar ventas por producto y tienda
ventas_por_producto = todas_tiendas.groupby(["Tienda", "Producto"]).size().reset_index(name="Ventas")

# Productos más vendidos por tienda (top 3)
top_productos = ventas_por_producto.sort_values(["Tienda", "Ventas"], ascending=[True, False])
top_3_por_tienda = top_productos.groupby("Tienda").head(3)

# Productos menos vendidos por tienda (últimos 3)
bottom_3_por_tienda = top_productos.groupby("Tienda").tail(3)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.barplot(
    data=top_3_por_tienda,
    x="Ventas",
    y="Producto",
    hue="Tienda",
    palette="viridis",
    dodge=True  # Para separar las barras por tienda
)
plt.title("Top 3 productos más vendidos por tienda")
plt.xlabel("Número de ventas")
plt.ylabel("Producto")
plt.legend(title="Tienda", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(
    data=bottom_3_por_tienda,
    x="Ventas",
    y="Producto",
    hue="Tienda",
    palette="magma",
    dodge=True
)
plt.title("Top 3 productos menos vendidos por tienda")
plt.xlabel("Número de ventas")
plt.ylabel("Producto")
plt.legend(title="Tienda", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
import pandas as pd

# Asignar nombres de tienda si aún no están definidos
tienda1["Tienda"] = "Tienda 1"
tienda2["Tienda"] = "Tienda 2"
tienda3["Tienda"] = "Tienda 3"
tienda4["Tienda"] = "Tienda 4"

# Unir todas las tiendas en un único DataFrame
todas_tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4])

# Calcular el costo de envío promedio por tienda (redondeado a 2 decimales)
costo_envio_promedio = todas_tiendas.groupby("Tienda")["Costo de envío"].mean().round(2).reset_index()
costo_envio_promedio.columns = ["Tienda", "Costo de Envío Promedio"]  # Renombrar columnas

# Ordenar de mayor a menor costo
costo_envio_promedio = costo_envio_promedio.sort_values("Costo de Envío Promedio", ascending=False)

print("Costo de envío promedio por tienda:")
print(costo_envio_promedio.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(
    data=costo_envio_promedio,
    x="Tienda",
    y="Costo de Envío Promedio",
    palette="Blues_r"  # Colores en degradé (azul)
)
plt.title("Costo Promedio de Envío por Tienda", fontweight="bold")
plt.xlabel("Tienda")
plt.ylabel("Costo Promedio (USD)")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)  # Líneas de referencia horizontales

# Añadir etiquetas con los valores
for index, row in costo_envio_promedio.iterrows():
    plt.text(index, row["Costo de Envío Promedio"], f"{row['Costo de Envío Promedio']:,.2f}",
             ha="center", va="bottom", fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Unir todas las tiendas y asegurar que las coordenadas sean válidas
todas_tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4])
todas_tiendas = todas_tiendas.dropna(subset=["lat", "lon"])  # Eliminar filas con coordenadas faltantes
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=todas_tiendas,
    x="lon",
    y="lat",
    hue="Tienda",
    palette="viridis",
    alpha=0.6,
    s=100  # Tamaño de los puntos
)
plt.title("Distribución Geográfica de Ventas por Tienda", fontweight="bold")
plt.xlabel("Longitud (lon)")
plt.ylabel("Latitud (lat)")
plt.legend(title="Tienda", bbox_to_anchor=(1.05, 1))
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
